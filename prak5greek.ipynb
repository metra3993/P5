На данной практической работе я буду разбирать датасет о "Греческие Боги". Он хранит в себе информацию об именах на разных языках (Греческий и Английский), их типах, подтипах и их описании. С помощью определённых функций узнаю ифнормацию о данном датасете и проденстрирую это на деле. 

```python
import pandas as pd
import numpy as np
greek_gods = pd.read_csv(r"C:\Users\user\Downloads\greek_gods.csv")
greek_gods.head()
```

Используя библиотеки Numpy и Pandas, а также файл CSV с датасетом при помощи read.csv и head мы выводим данный датасет.

```python
greek_gods.shape 
```

Размерность массива равна 445 строк на 5 столбцов, мэ это получили при помощи shape.

```python
greek_gods.columns 
```

При помощи columns мы выводим названия всех колонок нашего датасета. У нас присутсвует айдишник, имя на английском, имя на греческом, типах, описание.

```python
greek_gods.nunique()
```

Благодаря nunique мы получили все уникальные значения. 

```python
greek_gods.sort_values(by = 'main-type')
```

При помощи sort.values мы вывели все значения в порядке возрастания type 

```python
greek_gods['main-type'] = greek_gods['main-type'].replace('god' , 'боженька')
```

При помощи replace мы изменили все god в колонке main-type на Боженька

```python
greek_gods.drop_duplicates()
```

При помощи drop_duplicates мы удалили дубликаты.

```python
greek_gods.info() 
```

С помощью атрибута info мы получили общую информацию о датасете, включая количество непустых значений, типы данных каждого столбца, общее количество столбцов и количество использованной памяти. Мы видим, что всего 5 столбцов, количество использованной памяти = 17.5 + КB.

```python
greek_gods.describe()
```

При помощи describe мы вернули основные статистические показатели для каждого числового столбца, такие как количество, среднее значение, стандартное отклонение, минимальное и максимальное значение.

```python
greek_gods.loc[:, 'main-type' : 'description']
```

При помощи loc мы вывели столбцы с main-type до description.

```python
greek_gods.drop(columns = ['name-greek'])
```

При помощи drop мы удалили колонку name-greek. 

```python
greek_gods.to_csv (r"C:\Users\user\Downloads\greek_gods2.csv")
```

При помощи to_csv мы сохранили наш датасет на рабочий стол.
